```python
from scipy.stats import boxcox,kurtosis,skew

for i in continuous:
    print(i)
    print(df[i].skew() , df[i].kurtosis())
apply=[]


for i in continuous:
    if df[i].skew() <-0.7 or df[i].skew() >0.7 or df[i].kurtosis() >1 or df[i].kurtosis() <-1:
        apply.append(i)
print(apply)
```

```python
boxcox_cols = [
    'price', 'bathrooms', 'sqft_living', 'sqft_lot',
    'sqft_above', 'long', 'sqft_living15', 'sqft_lot15'
]

method=[0.5,-0.5,-1,1,2,-2]

for i in boxcox_cols:
    print(i)
    print("---------------")
    for j in method:
        print(j,end="\t")
        print("skew",pd.DataFrame(boxcox(df[i],lmbda=j)).skew().values,end="\t")
        print("kurtosis",pd.DataFrame(boxcox(df[i],lmbda=j)).kurtosis().values)
```

```python
df["price"]=boxcox(df["price"],lmbda=-0.5)
df["sqft_living"]=boxcox(df["sqft_living"],lmbda=0.5)
df["bathrooms"]=boxcox(df["bathrooms"],lmbda=0.5)
df["sqft_lot"]=boxcox(df["sqft_lot"],lmbda=0)
```

```python
#ML regression traning 
```

```python
x=df.drop("price",axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

```

```python
#Scalling
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error


model=LinearRegression()

model.fit(x_train_scale,y_train)
y_predict=model.predict(x_test_scale)
print("R2 Score:", r2_score(y_test, y_predict))
print("MAE:", mean_absolute_error(y_test, y_predict))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_predict)))
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor()
model.fit(x_train_scale,y_train)
y_predict=model.predict(x_test_scale)

print("R2 Score:", r2_score(y_test, y_predict))
print("MAE:", mean_absolute_error(y_test, y_predict))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_predict)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model=GradientBoostingRegressor()
model.fit(x_train_scale,y_train)
y_predict=model.predict(x_test_scale)

print("R2 Score:", r2_score(y_test, y_predict))
print("MAE:", mean_absolute_error(y_test, y_predict))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_predict)))
```

```python
import pickle 

with open("model.pkl","wb") as f:
    pickle.dump(model,f)

with open("scaler.pkl","wb") as f:
    pickle.dump(scaler,f)
```
